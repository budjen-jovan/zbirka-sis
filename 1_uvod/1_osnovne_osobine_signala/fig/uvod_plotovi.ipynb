```python
from pylab import *
import matplotlib as mpl
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif"
    })

plt.rc('text.latex', preamble=r'\usepackage[T1,T2A]{fontenc}\usepackage[utf8]{inputenc}\usepackage{cmsrb}\usepackage{amsmath}\usepackage[Symbolsmallscale]{upgreek}')

font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 10}

matplotlib.rc('font', **font)

def u(t):
    return (t > 0) * 1.0
```

```python
def generate_canvas(w = 1.6, a = 1.618):
    figure (figsize = (a*w, w), dpi = 100)
    xlim([-2,2])
    ylim([-2,2])
    minorticks_on()
    grid(b=True, which='major', color='0.45', linestyle='-')
    grid(b=True, which='minor', color='0.75', linestyle='--')


def plot_dirac(x, A, lw = 1.5, c = 'k', annotation = ''):
    #arrow(x, 0, 0, A, shape="full", color=c,linewidth=lw, head_width=0.2, head_length=0.3, length_includes_head=True, zorder=10)
    annotate(annotation, [x,A], [x,0], arrowprops=dict(width=lw, headwidth=10/1.618, headlength=10, color = c))
```

```python
generate_canvas( 1.5*1, 4/3 )
ylim([ -2,2 ])

x = linspace(-10, 10, 1000)

y = [ 0  for t in x]
plot(x, y, 'k', linewidth=1.5)
plot_dirac(+1, 1, lw = 1)
plot_dirac(-1, 1, lw = 1)

y = x**2 - 1
plot(x, y, 'k--', linewidth=2)

savefig("dirac_pm1.pdf", bbox_inches="tight", pad_inches=0)
```

```python
#plt.gray()

figure( figsize=(1.5*1.4, 1.5*1), dpi = 200) 


xlim([-2,6])
ylim([-2.5,2.5])


x = arange(-10, 10)


def y_fn(n):
    return u(n) - 2*u(n-4)

y = [ y_fn(n)  for n in x]
stem(x+0.05, y, use_line_collection=True, basefmt="k", markerfmt="k.", linefmt = 'k')


y2 = [y_fn(n) - y_fn(n-1) for n in x]
stem(x-0.05, y2, use_line_collection=True, basefmt="r", markerfmt="r.", linefmt='r' )

#title(u"(а)")

legend(["$x$", r"$\nabla x$"])

minorticks_on()
grid(b=True, which='major', color='0.85', linestyle='-')
grid(b=True, which='minor', color='0.75', linestyle='--')

savefig("crtaj_dt_a.pdf", bbox_inches="tight")
```

```python
#plt.gray()

figure( figsize=(1.5*1.4, 1.5*1), dpi = 200) 

xlim([-4,4])
ylim([-10,5])


x = arange(-10, 10)


def y_fn(n):
    if (n == -2):
        return n**2
    if (n == 2):
        return -2*n**2
    return 0

y = [ y_fn(n)  for n in x]
stem(x, y, use_line_collection=True, basefmt="k", markerfmt="k.", linefmt = 'k')


#y2 = [y_fn(n) - y_fn(n-1) for n in x]
stem(x+0.05, cumsum(y), use_line_collection=True, basefmt="r", markerfmt="r.", linefmt='r' )

legend(["$x$", r"$\nabla x$"])

minorticks_on()
grid(b=True, which='major', color='0.85', linestyle='-')
grid(b=True, which='minor', color='0.75', linestyle='--')

savefig("crtaj_dt_b.pdf", bbox_inches="tight")
```
